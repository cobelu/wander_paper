```python
%config Completer.use_jedi = False
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn')
```

## Load the data

```python
desc_file = 'index.csv'
runs_file = 'runs.csv'
dsgd_native_file = 'data/dsgdpp.csv'
netflix = 'data/netflix.npz'
yahoo = 'data/yahoo.npz'
movielens = 'data/movielens.npz'
```

```python
desc_df = pd.read_csv(desc_file)
runs_df = pd.read_csv(runs_file)
df = runs_df.merge(desc_df, on='run', how='left')
df['norm_time'] = df['time'] * df['n']
df = df.astype({'n': int})
```

```python
LFONT = 18

def plot_worker(df, method, title):
    worker_df = df.set_index('time')
    worker_df.groupby('n')['rmse'].plot(legend=True)
    plt.xlabel('Time (s)', weight='bold', fontsize=LFONT)
    plt.ylabel('RMSE', weight='bold', fontsize=LFONT)
    plt.legend(fontsize=LFONT)
    plt.savefig('scale_figs/{0}_{1}.pdf'.format(method, title))
#     plt.show()
    
def plot_norm_worker(df, title=''):
    worker_df = df.set_index('norm_time')
    worker_df.groupby('n')['rmse'].plot(legend=True)
    plt.title(title)
    plt.legend(fontsize=LFONT)
    plt.xlabel('Seconds â€¢ Workers')
    plt.ylabel('RMSE')
#     plt.show()
    
def plot_methods(df, method, name):
    worker_df = df.set_index('time')
    worker_df.groupby('method')['rmse'].plot(legend=True)
    plt.legend(fontsize=LFONT)
    plt.xlabel('Time (s)', weight='bold', fontsize=LFONT)
    plt.ylabel('RMSE', weight='bold', fontsize=LFONT)
    plt.savefig('comp_figs/{0}-{1}.pdf'.format(name, method))
#     plt.show()
    
def plot_nnz(df, method, name):
    worker_df = df.set_index('time')
    worker_df.groupby('method')['nnz'].plot(legend=True)
    plt.legend(fontsize=LFONT)
    plt.xlabel('Time (s)', weight='bold', fontsize=LFONT)
    plt.ylabel('Updates', weight='bold', fontsize=LFONT)
    plt.savefig('comp_figs/{0}-{1}.pdf'.format(name, method))
#     plt.show()
```
